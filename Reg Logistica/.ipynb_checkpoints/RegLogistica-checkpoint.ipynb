```python
DataRegLogistica[['glucose','prevalentHyp','age']].plot.scatter(x='glucose',y='prevalentHyp')
```

```python
# Defino 2 parametros optimos para la grafica
w = 0.09 # Pendiente
b = -15 # Intercepto
```

```python
# Agrego los puntos a la recta
x = np.linspace(0,DataRegLogistica['glucose'].max(),100)
y = 1/(1+np.exp(-(w*x+b)))

# Grafica de la recta
DataRegLogistica.plot.scatter(x='glucose', y='prevalentHyp')
plt.plot(x, y, '-r')
plt.ylim(0,DataRegLogistica['prevalentHyp'].max()*1.1)

plt.grid()
plt.show()
```
